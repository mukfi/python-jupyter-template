### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt
# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
# Create the scatter plot
plt.scatter(categories, values, color='b')
# Annotate each point with its (x, y) value
for i, (x, y) in enumerate(zip(categories, values)):
    plt.text(x, y, f'({x}, {y})', fontsize=9, ha='right')
# Customize the plot
plt.title('Scatter Plot with Annotations')  # Add title
# Display the plot
plt.show()
```
