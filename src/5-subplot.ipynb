### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.arange(0, 11)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
# Create the 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
# Plot a line chart in the first subplot
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x values')
axs[0, 0].set_ylabel('y values')
axs[0, 0].grid(True)
# Plot a bar chart in the second subplot
axs[0, 1].bar(categories, values, color=['r', 'g', 'b', 'y'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')
# Plot a scatter plot in the third subplot
axs[1, 0].scatter(categories, values, color='b')
for i, (x, y) in enumerate(zip(categories, values)):
    axs[1, 0].text(x, y, f'({x}, {y})', fontsize=9, ha='right')
axs[1, 0].set_title('Scatter Plot with Annotations')
# Plot a pie chart in the fourth subplot
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout
```
