### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt
# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
# Colors for each bar
colors = ['r', 'g', 'b', 'y']
# Create the bar plot
plt.bar(categories, values, color=colors)
# Customize the plot
plt.title('Bar Plot of Categories and Values')  # Add title
# Display the plot
plt.show()
```
